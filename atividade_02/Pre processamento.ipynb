# 3 Divisão treino/teste

## 3.4 Salvando os dados

```python
pickle.dump(x_train, open('X_train.pickle', 'wb'))
pickle.dump(x_test, open('X_test.pickle', 'wb'))
pickle.dump(y_train, open('y_train.pickle', 'wb'))
pickle.dump(y_test, open('y_test.pickle', 'wb'))
```

```python
def read_pickle(name):
    with (open(name, 'rb')) as openfile:
        while True:
            try:
                one_instance = pickle.load(openfile)
            except EOFError:
                break
    one_instance = np.asanyarray(one_instance)
    return one_instance
```

```python
x_train = read_pickle('X_train.pickle')
x_test = read_pickle('X_test.pickle')
y_train = read_pickle('y_train.pickle')
y_test = read_pickle('y_test.pickle')
```
